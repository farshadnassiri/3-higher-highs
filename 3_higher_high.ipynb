```python
import jdatetime
import datetime

# فرض کنید تاریخ میلادی دارید:
gregorian_date = datetime.datetime(2025, 10, 29)

# تبدیل به شمسی:
jalali_date = jdatetime.date.fromgregorian(date=gregorian_date.date())
print(jalali_date)
# خروجی مثلاً: 1404-08-18

# اگر بخواهی فقط چاپ زیباتر:
print(jalali_date.strftime("%Y/%m/%d"))  # 1404/08/18
```
